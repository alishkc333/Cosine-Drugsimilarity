```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
drug_feature=pd.read_csv("drug_feature.csv",index_col=0)
```

```python
df=drug_feature
```

```python
drug_names = df['drug']
features = df.drop(columns=['drug'])

cos_sim_matrix = cosine_similarity(features)

cos_sim_df = pd.DataFrame(cos_sim_matrix, index=drug_names, columns=drug_names)

```

```python
cos_sim_df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Reset index safely without column name conflicts
cos_sim_df_reset = cos_sim_df.copy()
cos_sim_df_reset.index.name = 'Drug1'
cos_sim_long = cos_sim_df_reset.reset_index().melt(id_vars='Drug1', var_name='Drug2', value_name='Similarity')

# Step 2: Remove self-pairs and duplicate pairs (A-B and B-A)
cos_sim_long = cos_sim_long[cos_sim_long['Drug1'] != cos_sim_long['Drug2']]
cos_sim_long['pair_key'] = cos_sim_long.apply(lambda row: tuple(sorted([row['Drug1'], row['Drug2']])), axis=1)
cos_sim_long = cos_sim_long.drop_duplicates(subset='pair_key')

# Step 3: Get top 10 most similar and dissimilar drug pairs
top_pos = cos_sim_long.sort_values(by='Similarity', ascending=False).head(10)
top_neg = cos_sim_long.sort_values(by='Similarity', ascending=True).head(10)
top_20 = pd.concat([top_pos, top_neg])

# Step 4: Create a symmetric heatmap matrix
heatmap_data = pd.pivot_table(top_20, index='Drug1', columns='Drug2', values='Similarity', fill_value=0)
heatmap_data = heatmap_data.combine_first(heatmap_data.T)  # Make it symmetric

# Step 5: Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', center=0, linewidths=0.5, linecolor='gray')
plt.title("Top 10 Positive and Negative Drug Cosine Similarities")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
